```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/sample_data/mnist_train_small.csv")
```

```python
X_train=data.iloc[:,1:]
y_train=data.iloc[:,0]
```

```python
X_train,y_train
```

```python
X_train= X_train/255.0
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(128,activation='relu'),
                           tf.keras.layers.Dense(10,activation='softmax')])
```

```python
model.compile(optimizer='adam',
  loss='sparse_categorical_crossentropy',
  metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=64,epochs=10)
```

```python
data_test=pd.read_csv('/content/sample_data/mnist_test.csv')
```

```python
data_test
```

```python
X_test=data.iloc[:,1:]
y_test=data.iloc[:,0]
```

```python
y_ans=model.predict(X_test)
```

```python
y_ans=[i.argmax() for i in y_ans]
```

```python
def accuracy(y_pred,y_true):
  return (y_true-y_pred)/len(y_pred)
```

```python
accuracy(y_ans,y_test)
```

```python
model.evaluate(X_test,y_test)
```
